```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.image as mpimg
import datetime
```

```python
# Cleaning Data
City_Pop_df = pd.read_csv('City Pop.csv')
City_Pop_df = City_Pop_df.replace(',', '', regex=True)
City_Pop_df['Pop'] = City_Pop_df['Pop'].astype(float)
# Plot Data
City_Pop_df.plot(kind='bar', x='City', y='Pop',figsize=(15,10), color='#4c74fb', legend=False)
plt.xlabel('City')
plt.ylabel('Population')
plt.show()
```
